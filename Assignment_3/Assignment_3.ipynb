# Basics
- Create two arrays: x, y  and z where:
    - x contains 12 evenly spaced numbers between 1 and 20 
    - y contains the values of x multiplied by 2 ie. 2*x
    - z contains the squares of elements of x ie x**2

- Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)

- Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.

- Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

- Finally, redo all the above plots using seaborn which is a higher level plotting library. It creates more aesthetically pleasing charts than matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
# - Create two arrays: x, y  and z where:
#    - x contains 12 evenly spaced numbers between 1 and 20 
#    - y contains the values of x multiplied by 2 ie. 2*x
#    - z contains the squares of elements of x ie x**2
x = np.linspace(1, 20, 12)
y = x * 2
z = x ** 2

print('X:',x)
print('Y:',y)
print('Z:',z)
```

# Matplotlib

```python
# Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 2x')

plt.xlim(0, 25)
plt.ylim(0, 45)
plt.show()
```

```python
# Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.
plt.plot(x, y, label="y=2x")
plt.plot(x, z, label="z=x^2")
plt.legend()
plt.show()
```

```python
# Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].plot(x, y)
axes[0].set_title("y=2x")
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x, z)
axes[1].set_title("z=x^2")
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

plt.show()
```

# Seaborn

```python
import seaborn as sns

sns.set_theme()

# Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)
line1 = sns.lineplot(x, y)
line1.set(xlabel='x', ylabel='y', xlim=(0, 25), ylim=(0, 45))
line1.set_title("y=2x")

plt.show()
```

```python
# Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.
sns.lineplot(x, y, label="y=2x")
sns.lineplot(x, z, label="z=x^2")
plt.show()
```

```python
# Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

line1 = sns.lineplot(x, y, ax=axes[0])
line1.set(xlabel='x', ylabel='y')
line1.set_title("y=2x")

line2 = sns.lineplot(x, z, ax=axes[1])
line2.set(xlabel='x', ylabel='z')
line2.set_title("z=x^2")

# axes[1].plot(x, z)
# axes[1].set_title("z=x^2")
# axes[1].set_xlabel('x')
# axes[1].set_ylabel('z')

plt.show()
```

- Create 3 arrays:

- Plot the scatterplot of x1 and x2. Color the points in the scatterplot according to their values in y: blue for 0 and red for 1. For example, color the point (5, 99) blue as the corresponding y value is 0 and the point (7, 86) red as the corresponding y value is 1.


```python
# Create 3 arrays:
#     - x1 = [5,7,8,7,2,17,2,9,4,11,12,9,6]
#     - x2 = [99,86,87,88,111,86,103,87,94,78,77,85,86]
#     - y = [0, 1, 1, 0, 0, 1, 0, 0, 1, 0, 0, 1, 1]

x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
x2 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
y = np.array([0, 1, 1, 0, 0, 1, 0, 0, 1, 0, 0, 1, 1])

print("x1", x1)
print("x2", x2)
print("y", y)
```

```python

# Plot the scatterplot of x1 and x2. Color the points in the scatterplot according to their values in y: blue for 0 and red for 1. For example, color the point (5, 99) blue as the corresponding y value is 0 and the point (7, 86) red as the corresponding y value is 1.
plt.scatter(x1, x2, c=['blue' if v  == 0 else 'red' for v in y])
plt.show()
```

    - Plot a bar plot of the ‘outcome’ variable. Is the dataset balanced or unbalanced? Try to explore the consequences of having such a dataset.
 
    - Create a figure containing 8 subplots arranged in (2, 4) grid. Plot the histograms of the 8 features (all the columns except the outcome) present in the dataset. Identify the features having normal distribution and skewed distributions.

    - Create a figure containing 8 subplots arranged in (4, 2) grid. Plot the boxplot for each of the features in each of the subplots. Outliers are a set of observations that vary significantly from the other observations. Which feature do you think has the most outliers? And Why?

    - A correlation matrix represents how much the features are correlated to each other. You can compute the correlation matrix of a dataframe using the dataframe.corr() method. Use matplotlib.pyplot.matshow to plot the correlation matrix in matplotlib. Also, you can use the matplotlib.pyplot.colorbar() to better explain your correlation matrix.


```python
import pandas as pd

df = pd.read_csv('pima_diabetes.csv')
df.head()
```

```python
#  Plot a bar plot of the ‘outcome’ variable. Is the dataset balanced or unbalanced? Try to explore the consequences of having such a dataset.

plt.bar(np.arange(len(df.Outcome)), df.Outcome, edgecolor='blue')
plt.show()

print('The dataset is clearly not balanced because there are a lot more values 1 than there are values 0')
```

```python
# Create a figure containing 8 subplots arranged in (2, 4) grid. Plot the histograms of the 8 features (all the columns except the outcome) present in the dataset. Identify the features having normal distribution and skewed distributions.
fig, ax = plt.subplots(2, 4,  figsize=(20, 10))

for i in range(0, 2):
    for j in range(0, 4):
        ax[i][j].hist(df.iloc[:, i*4+j], bins=10)
        ax[i][j].set_title(df.columns[i*4+j])
plt.show()

print("It appears that Glucose, Blood Pressure and BMI resemble a normal distribution while the rest resemble a skewed distribution.")
```

```python
# Create a figure containing 8 subplots arranged in (4, 2) grid. Plot the boxplot for each of the features in each of the subplots. Outliers are a set of observations that vary significantly from the other observations. Which feature do you think has the most outliers? And Why?
fig, ax = plt.subplots(4, 2,  figsize=(10, 20))

for i in range(0, 4):
    for j in range(0, 2):
        ax[i][j].boxplot(df.iloc[:, i*2+j])
        ax[i][j].set_title(df.columns[i*2+j])
plt.show()

print("It appears that Insulin and DiabetesPedigreeFunction have the most number of outliers. The fact is visible from the plot as there are a number of readings outside the whiskers of the plot")
```

```python
# A correlation matrix represents how much the features are correlated to each other. You can compute the correlation matrix of a dataframe using the dataframe.corr() method. Use matplotlib.pyplot.matshow to plot the correlation matrix in matplotlib. Also, you can use the matplotlib.pyplot.colorbar() to better explain your correlation matrix.

fig, ax = plt.subplots(1, 1, figsize=(15, 15))
corr = df.corr()
display(corr)
corr_plt = ax.matshow(corr)
plt.colorbar(corr_plt)

plt.show()
```
