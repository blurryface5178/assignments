Create a pandas series with some random numbers and texts and some null values.
- Add some data to your series
- Change data type to integer
- Remove the null values
- Sort the index of your series alphabetically 
- Count the occurrences of unique items of the series
- Sort the given series in descending order

```python
import numpy as np
import pandas as pd

# Create a pandas series with some random numbers and texts and some null values
series = pd.Series({"H":1,"O":2,"W":3,"A":"Apple","R":None,"E":"Banana","U":None})
print(series)
print()

# Add some data to your series
series = series.append(pd.Series({'D': 'Abhijan'}))
print(series)
print()

# # Change data type to integer
# col = series[0].astype('int32')
# print(col)
print('Returns an error because str and null cannot be cast to int32')
print()

# Remove the null values
series = series.dropna(0, how='any')
print(series)
print()

# Sort the index of your series alphabetically
series_sorted = series.sort_index()
print(series_sorted)
print()

# Count the occurrences of unique items of the series
uniques = series.value_counts()
print(uniques)
print()

# Sort the given series in descending order
desc = series.sort_index(ascending=False)
print(desc)
```

Create a dataframe having 5 columns and 50 random data

- Retrieve 1st and 5th column with some condition applied to 2nd column (Condition Example: all rows having value greater than x).
- Sample some of the data from the 1st column into a list and add some additional random data in it. Filter the dataframe and retrieve only those rows if 1st column contains the sampled data (Explore ‘isin’)

```python
# Create a dataframe having 5 columns and 50 random data
df = pd.DataFrame(np.random.randint(100, size=(50, 5)), columns=['A','B','C','D','E'])

# Retrieve the data from index 30-40 with the first 3 columns (Note: Use column index only)
selected_df_1 = df.iloc[30:40,:3]
print("First 3 cols, row 30-40")
display(selected_df_1)
print()

# Retrieve last 10 rows of 2nd and 5th columns (Note: Use column name only)
selected_df_2 = df[['B','E']].iloc[-10:]
print("2nd and 5th cols, last 10 rows")
display(selected_df_2)
print()

# Retrieve 1st and 5th column with some condition applied to 2nd column (Condition Example: all rows having value greater than x)
print("1st and 5th cols when val of col 2nd is greater than 50")
selected_df_3 = df[df['B']>50].iloc[:,[0,4]]
display(selected_df_3)
print()

# Sample some of the data from the 1st column into a list and add some additional random data in it.
selected_list = np.array([df.iloc[20:30,:1].values[x][0] for x in range (0,10)])
selected_list = np.append(selected_list, np.random.randint(100, size=(5, 1)))
# print(selected_list)
# print()

# Filter the dataframe and retrieve only those rows if 1st column contains the sampled data (Explore ‘isin’)
selected_df_4 = df[df['A'].isin(selected_list)]
print("All cols when rows of 1st col contains data from list", selected_list)
display(selected_df_4)

# Update the value of the 1st and 3rd column with new values in the 35th index of row
# display(df.loc[35,['A','C']])
df.at[35,['A', 'C']] = np.random.randint(100),np.random.randint(100)
# display(df.loc[35,['A','C']])

# Delete the row of the index position: 35, 39, 51. And reindex the dataframe
df =  df.drop([35, 39, 41])
df = df.reindex(np.arange(60))
display(df)
```

Create two dataframe (first dataframe having 50 random data and second dataframe having 40 random data). Each dataframe should have 3 columns including a common column (example: id) and 90% of data in common column should have identical data
- Merge both the dataframe using:
    - Inner
    - Outer
    - Left
    - Right
- Evaluate the differences observed 
- Use join() to combine the dataframe
- Use concat to combine the dataframe
- Evaluate the differences observed between merge, join and concat_

```python
# Create two dataframe (first dataframe having 50 random data and second dataframe having 40 random data). Each dataframe should have 3 columns including a common column (example: id) and 90% of data in common column should have identical data
df1 = pd.DataFrame(np.random.randint(100, size=(50, 2)), columns=['A','B'])
index  = pd.Series(np.arange(0, 50), name='id', dtype='int32')
df1 = pd.concat([index, df1], axis=1)
# df1.set_index(np.arange(0, 50))
df2 = pd.DataFrame(np.random.randint(100, size=(40, 2)), columns=['A','C'])
index  = pd.Series(np.arange(0, 40), name='id', dtype='int32')
df2 = pd.concat([index, df2], axis=1)

# display(df1)
# display(df2)

# Merge both the dataframe using:
#     - Inner
inner = df1.merge(df2, how='inner', on='id')
print("Inner")
display(inner)

#     - Outer
outer = df1.merge(df2, how='outer', on='id')
print("Outer")
display(outer)

#     - Left
left = df1.merge(df2, how='left', on='id')
print("Left")
display(left)

#     - Right
right = df1.merge(df2, how='right', on='id')
print("Right")
display(right)

# - Evaluate the differences observed 
print("Inner: takes intersection of keys of both dfs")
print("Outer: takes union of keys of both dfs")
print("Left: takes keys from left df")
print("Right: takes keys from right df")

# Use join() to combine the dataframe
joined = df1.join(df2, on="id", rsuffix='-r', how='outer')
print("Joined")
display(joined)

# - Use concat to combine the dataframe
concated = pd.concat([df1, df2],keys=['Series 1', 'Series 2'])
print("Concat")
display(concated)

# Evaluate the differences observed between merge, join and concat_
print("Merge: allows to merge two tables based on an index or a column. Possible use case: to add new columns to existing dataframe based on identifying column")
print("Join: is similar to merge but a little limited, allows to join two dataframes based on indices, Possible use case: to 'join' dataframes ")
print("Concat: allows to append values to existing dataframe. Possible use case: to add data to existing dataframe")
print()
print('Out of the three, merge is the most flexible and concat the most restrictive')
```

Create a dataframe having 3 columns: name, nationality, age and perform:
- Use filter() to get all data of column name and age
- Retrieve all the data of the column starting with “na”
- Retrieve all the data of the column ending with “e”
- Retrieve all the data of the column having subset “am” (hint: explore ‘like’ in filter())

```python
# Create a dataframe having 3 columns: name, nationality, age
names = pd.Series(['Abhijan', 'Prabesh', 'Muna','Sangam','Rohan','Mamata'])
age = pd.Series([50, 42, 23, 37, 28, 23])
nationality = pd.Series(['US','UK','Norway','Australia','Netherlands','Japan'])
df = pd.DataFrame({'Name':names,'Age':age,'Nationality':nationality})
# display(df)

# Use filter() to get all data of column name and age
name_and_age = df.filter(items=['Name', 'Age'])
print("Name and Age")
display(name_and_age)

# - Retrieve all the data of the column starting with “na”
start_Na = df.filter(regex='^(Na)')
print("Cols start with 'Na'")
display(start_na)

# - Retrieve all the data of the column ending with “e”
end_e = df.filter(regex='(e)$')
print("Cols end with 'e'")
display(end_e)

# - Retrieve all the data of the column having subset “am” (hint: explore ‘like’ in filter())
have_am = df.filter(like='am')
print("Cols containing 'am'")
display(have_am)
```

- Use this dataset and perform the following tasks:
    - Read the dataset as pandas dataframe
    - Find the number of cases for each country
    - Filter the Country_Region whose value == “India” and assign it in dataframe df_india.
        - Retrieve the active cases of “Madhya Pradesh” Province from df_india.
        - Retrieve the state with the maximum number of confirmed cases from df_india.
    - Filter the Country_Region whose value == “Canada” and assign it in dataframe df_canada.
        - Province/State Diamond Princess  has the “NaN” value for Active number of cases. Fill it with the median of the Active cases.
    - Filter the Country_Region whose value == “US” and assign it in dataframe df_us.
        - Groupby the df_us dataframe by ‘State_Province’ and aggregate with maximum values of the attributes and assign it in df_us_states.
        - Re-index the dataframe df_us_states.
        - The United States has 50 states. df_us_states contains additional states which are not the states of the United States. Remove those additional states from the dataframe.
        - Which state has the minimum number of active cases?

```python
# Read the dataset as pandas dataframe
df = pd.read_csv('dataset.csv')
df.head()

# Find the number of cases for each country
cases = df.filter(items=['Country_Region', 'Confirmed']).groupby(['Country_Region']).sum()
cases.head()
```

```python
# Filter the Country_Region whose value == “India” and assign it in dataframe df_india
df_india = df[df['Country_Region']=='India']
display(df_india)

# Retrieve the active cases of “Madhya Pradesh” Province from df_india.
madhya_pradesh = df_india[df_india['Province_State']=='Madhya Pradesh'].Active
print("Active Cases in Madhya Pradesh",madhya_pradesh.values[0])

# Retrieve the state with the maximum number of confirmed cases from df_india.
id = df_india['Confirmed'].idxmax()
max_india = df_india.loc[id].Province_State
print("Maximum Confirmed Cases in state:",max_india)
```

```python
# Filter the Country_Region whose value == “Canada” and assign it in dataframe df_canada.
df_canada = df[df['Country_Region']=='Canada']

# - Province/State Diamond Princess  has the “NaN” value for Active number of cases. Fill it with the median of the Active cases.
df_canada.at[df_canada.Province_State=="Diamond Princess", 'Active'] = df_canada.Active.median()
print("Filled median value",df_canada.Active.median(), "for Diamond Princess")
display(df_canada)
```

```python
# - Filter the Country_Region whose value == “US” and assign it in dataframe df_us.
df_us = df[df['Country_Region']=='US']

# - Groupby the df_us dataframe by ‘State_Province’ and aggregate with maximum values of the attributes and assign it in df_us_states.
df_us_states = df_us.groupby('Province_State').max()
print("Grouped by 'Province_State'")
display(df_us_states)

# - Re-index the dataframe df_us_states.
df_us_states = df_us_states.reset_index().set_index(np.arange(len(df_us_states.index)))
print("Re-indexed")
display(df_us_states)

# - The United States has 50 states. df_us_states contains additional states which are not the states of the United States. Remove those additional states from the dataframe.
clean_us_states = df_us_states[df_us_states['Admin2'].isnull() != True]
clean_us_states = clean_us_states[(clean_us_states['Province_State'].str.contains('District of Columbia') != True)]
clean_us_states = clean_us_states[(clean_us_states['Province_State'].str.contains('Puerto Rico') != True)]
print("Number of entries after cleaning is: ",len(clean_us_states.index))
display(clean_us_states)


# - Which state has the minimum number of active cases?
min_state = clean_us_states.loc[clean_us_states.Active.idxmin()].Province_State
print(min_state)
```
